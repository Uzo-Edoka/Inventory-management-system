### Build an inventory management system for a supermarket

```python
import pandas as pd
```

```python
def run_inventory_ms():
    while True:
        print('\nInventory management system')
        print('1. Add Product')
        print('2. View Inventory')
        print('3. Search Product')
        print('4. Delete Product')
        print('5. Exit')

        choice = input('Enter your choice (1-5): ')
        if choice == '1':
            add_product()
        elif choice == '2':
            view_inventory()
        elif choice == '3':
            search_product()
        elif choice == '4':
            delete_product()
        elif choice == '5':
            break
        else: 
            print('Invalid choice. Please enter a number between 1 and 5.')
```

```python
# add product function
def add_product():
    product_name = input('Enter the product name: ')
    
    try:
        # read our csv file to pandas dataframe
        df = pd.read_csv('inventory_data.csv') # removed header & column, header=None, names=['Product Name', 'Quantity', 'Price'])
        
        # check if product already exists
        existing_product = df[df['Product Name'].str.lower() == product_name.lower()]
        
        if not existing_product.empty:
            print(f'Product: {product_name} already exists in inventory.')
            update_option = input('Do you want to update the existing product? (yes/no): ').lower()
            
            if update_option == 'yes':
                # get new quantity and price
                new_quantity = int(input('Enter new quantity: '))
                new_price = float(input('Enter new price: '))
                
                # .loc to locate the product that has the product name
                df.loc[df['Product Name'].str.lower() == product_name.lower(), 
                       ['Quantity', 'Price']] = [new_quantity, new_price]
    
                # write the update from df to the csv file
                df.to_csv('inventory_data.csv', index=False)
                print(f'Product: {product_name} updated successfully.')
                return
            else:     #  if update_option == 'no'
                print('No changes made. ')
                return
    except FileNotFoundError:   # if the product does not already exist in our file
         # create new dataframe
        df = pd.DataFrame(columns = ['Product Name', 'Quantity', 'Price'])
        # get values for new product
    quantity = int(input('Enter the quantity: '))
    price = float(input('Enter the price per unit: '))
    
    # add products to df
    new_product = pd.DataFrame({'Product Name': [product_name], 'Quantity': [quantity], 'Price': [price]})
    df = pd.concat([df, new_product], ignore_index = True)
    
    # write the df to the csv file
    df.to_csv('inventory_data.csv', index=False)
    print(f'Product {product_name} added successfully!\n')
    
# function to view the inventory
def view_inventory():
    try: 
        df = pd.read_csv('inventory_data.csv') # removed the ff header=None, names=['Product Name', 'Quantity', 'Price'] 
                                            # to resolve repetitve creation of headers 
        if df.empty: 
            print('No products in the inventory')
        else:
            print('\n')
            print('Product Inventory: ')
            print(df)
    except FileNotFoundError:
        print('File does not exist')

# function to search for products

def search_product():
    search_name = input("Enter the name of the product to search: ")

    try: 
        df = pd.read_csv('inventory_data.csv') # removed header & columns, header=None, names=['Product Name', 'Quantity', 'Price'])

        result = df[df['Product Name'].str.lower() == search_name.lower()]

        if result.empty:
            print(f"No product with the name: '{search_name}'" )
        else:
            print('product found')
            return result
    except FileNotFoundError:
        print('File does not exist')

        # delete product
def delete_product():
    product_name = input(f'Enter the name of the Product: ')
    try:
        # read our csv file to pandas dataframe
        df = pd.read_csv('inventory_data.csv') # removed header & column, header=None, names=['Product Name', 'Quantity', 'Price'])
        
        # check if product already exists
        result = df[df['Product Name'].str.lower() == product_name.lower()]
        if result.empty:
            print(f"No product found with the name '{product_name}'.")
        else:
            df = df[df['Product Name'].str.lower() != product_name.lower()]
            df.to_csv('inventory_data.csv', index=False)
            
            # notify user that product has been deleted
            print(f"Product '{product_name}' deleted successfully.")
    except FileNotFound:
        print(f"Product '{product_name}' is not in the inventory.")
            
```
